```python
## training the model

## since we working on an unsuprevised algorithm , we will not be doing a train test split as we dont have the paramater labels


## import the k mean algorithm

from sklearn.cluster import KMeans

kmeans= KMeans(n_clusters=2)

## fit the model

kmeans.fit(df.drop('Private',axis=1))



```

```python
##center clusters
kmeans.cluster_centers_
```
